## User Evaluation

In this notebook, we randomly choose 5 cases from the seed case base in order to run a user evaluation. For each of those cases, we will print the image (xAI graph to explain) and two texts (the 'qwen_insight' zero-shot baseline and the Pixtral 'rag_solution' insights). We selected Qwen as the baseline for this comparison because it was the clear winner across almost every metric in the offline evaluation,  followed by our RAG output. 

We will attach each image and the corresponding two texts onto a Google Form, aiming to see if human preference aligns with the quantitative metrics obtained during offline evaluation by asking users to choose the the most useful option based on their criteria.

```python
import os
import pandas as pd
from IPython.display import display, Markdown, Image

# Path Setup
SRC_DIR = os.getcwd()
EVAL_ROOT = os.path.dirname(SRC_DIR)
PROJECT_ROOT = os.path.dirname(EVAL_ROOT)
XAI_RAG_DIR = os.path.join(PROJECT_ROOT, "xai_rag_cbr")
CASE_BASE_CSV = os.path.join(XAI_RAG_DIR, "data", "processed", "case_base.csv")
IMAGES_DIR = os.path.join(XAI_RAG_DIR, "data", "raw", "xai_images")
RAG_OUTPUT_CSV = os.path.join(EVAL_ROOT, "data", "raw", "pixtral_rag_insights.csv")

# Load and merge Data
df_base = pd.read_csv(CASE_BASE_CSV, dtype={'id': str})
df_base['id'] = df_base['id'].str.zfill(6)
df_rag = pd.read_csv(RAG_OUTPUT_CSV, dtype={'id': str})
df_rag['id'] = df_rag['id'].str.zfill(6)
df_master = pd.merge(df_base, df_rag, on="id", how="inner")

# Select 5 Random Cases (we will use a fixed seed for replicabilty of choice)
RANDOM_SEED = 23
df_sample = df_master.sample(n=5, random_state=RANDOM_SEED)

for index, row in df_sample.iterrows():

    case_id = row['id']

    # Get the image path
    img_path = os.path.join(IMAGES_DIR, f"{case_id}.png")
    
    # Extract the texts
    qwen_text = row['qwen_insight']   # Zero-shot Qwen baseline
    rag_text = row['insight']         # Few-shot RAG from Pixtral
    
    print(f"==============================================================================================================================================")

    print(f"CASE ID: {case_id}")

    if os.path.exists(img_path):
        display(Image(filename=img_path, width=800))
    else:
        print(f"[Warning: Image not found at {img_path}]")
    
    markdown_content = f"""
**>>> OPTION A: ZERO-SHOT QWEN BASELINE**
> {qwen_text}

**>>> OPTION B: PIXTRAL FEW-SHOT RAG**
> {rag_text}
<br>
"""
    display(Markdown(markdown_content))
```
